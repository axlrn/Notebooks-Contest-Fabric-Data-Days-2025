# Load datasets

```python
import kagglehub
from pathlib import Path
import shutil
import pandas as pd
from IPython.display import display

# Download
ds1 = kagglehub.dataset_download("programmer3/sustainable-urban-planning-and-landscape-dataset")
print("ds1 (retorno kagglehub):", ds1)

src = Path(ds1)
print("src (Path):", src)

# Localizar arquivo CSV real (se o download criou uma subpasta como '1')
if src.is_dir():
    csvs = list(src.rglob("*.csv"))
    if not csvs:
        print("Nenhum CSV encontrado em:", src)
        file_src = None
    else:
        file_src = csvs[0]
        print("CSV encontrado em:", file_src)
else:
    file_src = src if src.suffix.lower() == ".csv" else None

if not file_src:
    raise FileNotFoundError("Não foi possível localizar o arquivo .csv no resultado do download.")

# destino desejado (na raiz datasets)
dst = Path(".") / "datasets" / file_src.name
dst.parent.mkdir(parents=True, exist_ok=True)

# sobrescrever se necessário
if dst.exists():
    print("Destino já existe — será substituído:", dst)
    dst.unlink()

shutil.move(str(file_src), str(dst))
print("Movido para:", dst)

# mostrar preview se for CSV
if dst.suffix.lower() == ".csv":
    df = pd.read_csv(dst)
    display(df.head())
```
